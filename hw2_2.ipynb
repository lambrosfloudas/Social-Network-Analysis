```python
def exercise2(Graph):
    print("CNM:")
    print("---------------- ------------------ -----------------")
    t0 = time()
    modularity, CmtyV = Graph.CommunityCNM()
    for Cmty in CmtyV:
        print("Community: ")
        for NI in Cmty:
            print(NI)
    print("---------------- ------------------ -----------------")
    print("The modularity of the network is %f" % modularity)
    print("---------------- ------------------ -----------------")
    t1 = time()
    print('CNM takes %f seconds.'%(t1-t0))
    print("---------------- ------------------ -----------------")
    print("Girvan Newman:")
    print("---------------- ------------------ -----------------")
    t2 = time()
    modularity, CmtyV = Graph.CommunityGirvanNewman()
    for Cmty in CmtyV:
        print("Community: ")
        for NI in Cmty:
            print(NI)
    print("---------------- ------------------ -----------------")
    print("The modularity of the network is %f" % modularity)
    print("---------------- ------------------ -----------------")
    t3 = time()
    print('Girvan Newman takes %f seconds.'%(t3-t2))
```

```python
exercise2(G50)
```
